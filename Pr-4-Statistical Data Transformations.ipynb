```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import FunctionTransformer
```

```python
#step1
data=np.array([1,2,3,4,5,6],dtype=float)
```

```python
#logarithmic
transform=FunctionTransformer(func=np.log1p)
log_transformed_data=transform.fit_transform(data)
```

```python
#square
s_t_d=np.square(data)
```

```python
#reciprocol
r_t_d=np.reciprocal(data)
```

```python
#trignometric
sin_t_d=np.sin(data)
cos_t_d=np.cos(data)
tan_t_d=np.tan(data)
```

```python
#sqrt 
sqrt_t_d=np.sqrt(data)
```

```python
#combine
results=pd.DataFrame({
    'Orginal':data,
    'Log(log1p)':log_transformed_data,
    'Square':s_t_d,
    'Reciprocal':r_t_d,
    'Sin':sin_t_d,
    'Cos':cos_t_d,
    'Tan':tan_t_d,
    'Square root':sqrt_t_d
})

print(results)
```
